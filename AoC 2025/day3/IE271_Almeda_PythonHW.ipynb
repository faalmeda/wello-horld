```python
# Day 3

# Converts the txt file into a list of banks
with open('adventday3.txt', 'r') as f:
    day3_list = f.read().splitlines()

print(day3_list[0:3])
```

```python
# Part 1

# Function that parses each bank for the first and second highest digits
# Function returns the two digits as int
def get_jolt(x):
  jolt = 0

  if x:
    for i in x:

      s = str(x)
      n = len(x)
      highest_val = max(s)
      idx1 = s.index(highest_val)

    # Special case when highest digit is at the end
    if idx1 == n - 1:
      jolt += int(highest_val) # Ones digit
      s = s[ : n - 1]
      next_val = int(max(s))
      jolt += (next_val * 10) # Tens digit

    else:
      jolt += int(highest_val) * 10 # Tens digit
      s = s[idx1 + 1:]
      next_val = int(max(s))
      jolt += next_val # Ones digit

    return jolt

# Main def
def main1(banks):
  joltage = 0
  if banks:
    for battery in banks:
      joltage += get_jolt(battery)

  return joltage

print(f'Total joltage: {main1(day3_list)}')

```

```python
# Part 2

def get_jolt_2(x, no_battery=12, debug=True):

  # Check if x exists
  if x:
      s = str(x)
  else:
      return 0

  collected_batteries = []  # List to store the digits we find
  current_s = s             # The 'remaining' string we can search in

  if debug: print(f'Starting Scan on: {s}')

  for i in range(no_battery):
      batteries_needed_after_this = (no_battery - 1) - i

      # Reserve the last batteries depending on how many is still needed

      end = len(current_s) - batteries_needed_after_this

      # The slice of the bank we are allowed to touch
      search_window = current_s[:end]

      if debug:
          print(f'\n=== Battery {i+1} ===')
          print(f'Current Bank: {current_s}')
          print(f'Safety Window: {search_window} (Saving {batteries_needed_after_this} digits)')

      # Find the max digit in the safe window
      # Also require the idx to slice
      highest_val = '0'
      highest_idx = -1

      # Largest digit finder
      for idx, char in enumerate(search_window):
          if char > highest_val:
              highest_val = char
              highest_idx = idx
          if char == '9': # Max integer
              break

      if debug: print(f'Found max {highest_val} at relative index {highest_idx}')

      collected_batteries.append(highest_val)

      # Discard everything before and including the found digit
      current_s = current_s[highest_idx + 1:]

  # Join the collected digits and return integer
  final_result = ''.join(collected_batteries)

  if debug: print(f'\nFinal Result String: {final_result}')

  return int(final_result) if final_result else 0

# Testing
print(f'\nFinal jolt: {get_jolt_2(day3_list[34],12,True)}')
```

```python
def main2(banks):
  if banks:
    joltage = 0
    for battery in banks:
      joltage += get_jolt_2(battery,12,False)
    return joltage
  else:
    return 0

print(f'Total joltage: {main2(day3_list)}')
```
